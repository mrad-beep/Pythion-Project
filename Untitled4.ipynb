```python
#importing librarie:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# : Load the data forl data cleaning
# 
data = pd.read_csv('Heart.csv')

#inspecting the first Five rows 
data.head()

#inspect last 5 rows of our dataset 
data.tail()


#Shape of our dataset 
data.shape

print("Number of Rows", data.shape[0])
print("Number of columns", data.shape[1])

#information of our dataset
data.info()

#check null values
data.isnull().sum()

#check duplicate data and eliminate them
data = data.drop_duplicates()
data.shape

#overall statistics
data.describe()

#Draw Correlation matrix
plt.figure(figsize = (17,6))
sns.heatmap(data.corr(), annot=True)

#check how many persons have heart diseases and how many have not
data['target'].value_counts()
sns.countplot(data['target'])

#finding of no. of males and females in dataset

data['sex'].value_counts()


sns.countplot(data['sex'])
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()

#Gemder distribution 
sns.countplot(x = "sex", hue = "target", data = data)
plt.xticks([1, 0], ["Male", "Female"])
plt.legend(labels=["No-Diesease", "Disease"])
plt.show()

#age distribution in dataset
sns.displot(data['age'], bins=20)
plt.show()

#chest pain distribution
sns.countplot(x="cp", hue = "target", data=data)
plt.show()
```
